```python
import pandas as pd
import numpy as np

data_set = pd.read_csv('airbnb_lisbon_1480_2017-07-27.csv')
data_set.head()

data_set2 = data_set.apply(lambda x: x.astype(str).str.lower()) #this converts dataframe to lowercase
col_to_drop = ["country", "borough", "room_id", "survey_id", "host_id", "bathrooms", "minstay", "last_modified","latitude","longitude","location"]
my_data = data_set2.drop(col_to_drop, axis=1)
my_data[['reviews','overall_satisfaction','price']] = my_data[['reviews','overall_satisfaction','price']].astype('float')
#my_data.dtypes


#dropped_col_data = dropped_col_data[dropped_col_data.reviews !=0]
new_data = my_data[my_data["reviews"] != 0]

new_data.tail(20)
new_data.to_csv('cleaned_data.csv', index=False)
```

```python
# What is the relationship between neighbourhood and room type?
# Count number of room types per neighbourhood 

room_type = pd.get_dummies(new_data['room_type'])
```

```python
new_data = pd.concat([new_data, room_type], axis=1)
sorted_data = new_data.sort_values(by=['neighborhood'])
sorted_data.head()
```

```python
final_data = sorted_data.groupby('neighborhood').mean()
col_to_drop2 = ["reviews", "overall_satisfaction", "price"]
final_data = final_data.drop(col_to_drop2, axis=1)
```

```python
most_entire_home = final_data.sort_values(by=['entire home/apt'], ascending=False)
most_entire_home['entire home/apt'] = pd.Series(["{0:.2f}%".format(val * 100) 
                                                for val in most_entire_home['entire home/apt']], 
                                                index = most_entire_home.index)
most_entire_home['private room'] = pd.Series(["{0:.2f}%".format(val * 100) 
                                                for val in most_entire_home['private room']], 
                                                index = most_entire_home.index)
most_entire_home['shared room'] = pd.Series(["{0:.2f}%".format(val * 100) 
                                                for val in most_entire_home['shared room']], 
                                                index = most_entire_home.index)

most_entire_home.head()
```
